````

def iconfianza(df, confianza = 0.95):
  'Función para calcular intervalos de confianza de variables en un DataFrame'
  n = len(df)
  m = df.mean()
  q1 = df.quantile(0.25)
  q3 = df.quantile(0.75)
  iqr = q3 - q1
  std = df.std()
  h = std / math.sqrt(n) * stats.t.ppf((1 + confianza ) / 2, n -1)
  print('El 95% de los datos del Dataframe ingresado se encuentra entre los valores', m - h, m + h)
  print('El 95% de los datos del Dataframe con IQR se encuentra entre los valores', (q1 - 1.5 * iqr), (q3 + 1.5 * iqr))
  return;
````
